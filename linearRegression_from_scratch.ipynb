```python
import numpy as np
import matplotlib.pyplot as plt

# Load data
data = np.loadtxt("data.txt", delimiter=",")
X = data[:, 0]
Y = data[:, 1].reshape(X.size, 1)
X = np.vstack((np.ones((X.size, )), X)).T

# Scatter plot of data points
plt.scatter(X[:, 1], Y, label="Data Points")

def model(X, Y, learning_rate, iteration):
    m = Y.size
    w = np.zeros((2, 1))  # Replaced 'theta' with 'w'
    cost_list = []
    for i in range(iteration):
        y_pred = np.dot(X, w)
        cost = (1/(2*m)) * np.sum(np.square(y_pred - Y))
        d_w = (1/m) * np.dot(X.T, y_pred - Y)
        w = w - learning_rate * d_w
        cost_list.append(cost)
    return w, cost_list

iteration = 100
learning_rate = 0.00000005
w, cost_list = model(X, Y, learning_rate=learning_rate, iteration=iteration)

# Plot the regression line
x_values = np.linspace(min(X[:, 1]), max(X[:, 1]), 100)
y_values = w[0] + w[1] * x_values
plt.plot(x_values, y_values, color='red', label="Regression Line: y = {:.2f}x + {:.2f}".format(w[1, 0], w[0, 0]))

plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Linear Regression using Gradient Descent")
plt.grid(True)

# Add cost vs iterations plot
plt.figure()
plt.plot(range(iteration), cost_list)
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.title("Cost vs. Iterations")

plt.show()

print("Slope (m):", w[1, 0])
print("Y-intercept (b):", w[0, 0])
```
